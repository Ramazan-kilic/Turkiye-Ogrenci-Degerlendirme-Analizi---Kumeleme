* attendance: Devam seviyesinin kodu; {0, 1, 2, 3, 4} değerlerini alır
*  difficulty: Öğrencinin algısına göre dersin zorluk seviyesi; {1,2,3,4,5} değerlerini alır

*  Q1: Dönem başında ders içeriği, öğretim yöntemi ve değerlendirme sistemi sağlandı.

*  Q6: Ders kitabı ve diğer ders kaynakları yeterli ve günceldi.

*   Q7: Ders, saha çalışmaları, uygulamalar, laboratuvar, tartışmalar ve diğer çalışmalara olanak sağladı.

*   Q9: Dersten büyük keyif aldım ve ders sırasında aktif katılım konusunda hevesliydim.

*  Q10: Dönem veya yıl sonunda derse dair ilk beklentilerim karşılandı.

*   Q11: Ders, mesleki gelişimime uygun ve faydalıydı.


*  Q12: Ders, hayata ve dünyaya yeni bir bakış açısıyla bakmamı sağladı.


*  Q13: Öğretim elemanının bilgisi güncel ve ilgiliydi.


*   Q14: Öğretim elemanı derslere hazırlıklı geldi.


*   Q15: Öğretim elemanı duyurulan ders planına uygun olarak ders verdi.


*   Q16: Öğretim elemanı derse bağlıydı ve anlaşılırdı.

*   Q26: Öğretim elemanının değerlendirme sistemi (vize ve final soruları, projeler, ödevler vb.) dersin hedeflerini etkili bir şekilde ölçtü.
*   Q27: Öğretim elemanı sınavlara çözümler sağladı ve bunları öğrencilerle tartıştı.

**Temel bileşenler Analizi**

---



```python
X = df.iloc[:, 5:33]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)
X_pca
```

```python
# veri kümesinden ne kadar bilgi tuttuk
pca.explained_variance_ratio_.cumsum()[1]
```

**Model Eğitimi**

---



```python
from sklearn.cluster import KMeans
distortions = []
cluster_range = range(1,6)

for i in cluster_range:
    model = KMeans(n_clusters=i, init='k-means++', random_state=42)
    model.fit(X_pca)
    distortions.append(model.inertia_)
```

```python
plt.plot(cluster_range, distortions, marker='o')
plt.xlabel("Number of clusters")
plt.ylabel('Distortions')
plt.show()
```

```python
model = KMeans(n_clusters=3, init='k-means++', random_state=42)
model.fit(X_pca)
y = model.predict(X_pca)
```

```python
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], s=50, c='red', label='cluster 1')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], s=50, c='yellow', label='cluster 2')
plt.scatter(X_pca[y==2, 0], X_pca[y==2, 1], s=50, c='green', label='cluster 3')
plt.scatter(model.cluster_centers_[:,0], model.cluster_centers_[:, 1], s=100, c='blue', label='centroids')
plt.title('Cluster of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```

```python
from collections import Counter
Counter(y)
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
y = model.fit_predict(X_pca)
```

```python
plt.scatter(X_pca[y==0, 0], X_pca[y==0, 1], s=50, c='red', label='cluster 1')
plt.scatter(X_pca[y==1, 0], X_pca[y==1, 1], s=50, c='yellow', label='cluster 2')
plt.title('Cluster of students')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend()
plt.show()
```
